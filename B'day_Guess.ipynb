```python
# BirthDay Guesser

# Hey Guys. This was one of my first projects when I picked Programming. Uses basic python and a simple binary logic.
# Would love others to try this aswell and help me simplify the code. This gives a good exposure to basic python concepts.

# The user is shown a series of Cards each of a specific No. associated to them. These cards contain certain dates.
# The user is then asked to answer which of the cards by contain his/her Birthday. The user enters the no.s of each of the cards and voila, their b'day is shown!!!


# A greeting and intro...


print("Hello! This is a game to predict you Birthday.")  # print() function
name = input("What is you name?\n")  # taking user inputs in python
print("Hello ", name)  # using variables
print("Now I'll show you some cards numbered 0 to 4 which contain dates")

# Lets generate those cards...

cards = [[], [], [], [], []]  # list of lists
a = []
for i in range(1, 32, 1):  # for loop, in, range function
    a.append([int(x) for x in bin(i + 32)[3:]])  # changing to binary form and slicing of list

# arranging the dates into the cards ...

k = 0
for i in range(0, 31, 1):  # nested for loop
    k += 1
    for j in range(0, 5, 1):
        if a[i][j] == 1 and k not in cards[4 - j]:  # my cards are numbered backwards
            cards[4 - j].append(k)
        else:
            continue

        # labelling the cards 0 to 4 ...

k = 1
for i in cards:
    i.append("<<<---")
    i.append(k)
    k = k + 1

# Displaying the Card to the user...

print(*cards, sep='\n')

# The user now has to enter the no. of card which contain the date they are thinking...

print("\nYou have to enter the card number one by one which have the date of you birthday seperated by space\n\n")
print("\nWhich among 1,2,3,4,5 cards above has your B'day on it?")
x = [int(x) for x in input("Enter CARD NO.s that have your Birthday: ").split()]  # taking and storing multiple inputs

# B'day calculation logic
bday = 0
for i in x:
    bday = bday + cards[i - 1][0]

# Displaying the B'day

if ((bday % 10 in [1, 2, 3]) and (bday not in [11, 12, 13])):  # code to add 'st','nd','rd','th' at the desirable places

    if bday % 10 == 1:
        str1 = "st"
    elif bday % 10 == 2:
        str1 = 'nd'
    else:
        str1 = 'rd'

    print("Your Birthday is on " + str(bday) + str1 + " of the month")


else:
    print("Your Birthday is on " + str(bday) + 'th of the month')

```
